```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('../../dataset/meal_service/train.csv')
```

```python
breakfast = []
breakfast_cnt = {}
for row in df_train['조식메뉴']:
    # first, split by space
    splits = row.split(' ')
    menu = []
    for dish in splits:
        if len(dish) > 1:
            ## find (New)
            if '(New)' in dish:
                dish_str = dish.split('(New)')[0] + dish.split('(New)')[1]
            elif '(' not in dish and ')' not in dish:
                dish_str = dish
            menu.append(dish_str)
            
            # for counting menu
            if dish_str in breakfast_cnt:
                breakfast_cnt[dish_str] = breakfast_cnt[dish_str] + 1
            else:
                breakfast_cnt[dish_str] = 1
    if len(menu) > 3:
        breakfast.append(menu)
    else:
        # no meal today
        breakfast.append(['-'])
        
for menu in breakfast:
    print(menu)
            
            
```
