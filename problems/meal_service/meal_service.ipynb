```python
breakfast_menus = []

for menu in df_train['조식메뉴']:
    menu_split = menu.split(' ')

    menu_list = []
    for dish in menu_split:
        if len(dish) > 1:
            if dish[1] == 'N':
                # (New) 제거
                menu_list.append(dish.split(')')[1])
            elif dish[0] != '(' and dish[-1] != ')' :
                # 원산지 제거
                menu_list.append(dish)
    if len(menu_list) > 3:
        breakfast_menus.append(menu_list) 
    else:
        breakfast_menus.append(['*'])
```

```python
lunch_menus = []

for menu in df_train['중식메뉴']:
    menu_split = menu.split(' ')

    menu_list = []
    for dish in menu_split:
        if len(dish) > 1:
            if dish[1] == 'N':
                # (New) 제거
                menu_list.append(dish.split(')')[1])
            elif dish[0] != '(' and dish[-1] != ')' :
                # 원산지 제거
                menu_list.append(dish)
    if len(menu_list) > 3:
        lunch_menus.append(menu_list) 
    else:
        lunch_menus.append(['*'])
```

```python
dinner_menus = []

for menu in df_train['석식메뉴']:
    menu_split = menu.split(' ')

    menu_list = []
    for dish in menu_split:
        if len(dish) > 1:
            if dish[1] == 'N':
                # (New) 제거
                menu_list.append(dish.split(')')[1])
            elif dish[0] != '(' and dish[-1] != ')' :
                # 원산지 제거
                menu_list.append(dish)
    if len(menu_list) > 3:
        dinner_menus.append(menu_list) 
    else:
        dinner_menus.append(['*'])
```

```python
dish_cnt = {}
for menu in dinner_menus:
    if len(menu) == 1:
        continue
    else:
        if menu[1] not in dish_cnt:
            dish_cnt[menu[1]] = 1
        else:
            dish_cnt[menu[1]] = dish_cnt[menu[1]] + 1
print(dish_cnt)
        
```
